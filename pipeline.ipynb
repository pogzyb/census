```python
# data prep
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.metrics import roc_auc_score, recall_score, confusion_matrix

import xgboost as xgb
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression

# visualizations
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import gc
```

```python
# read in data

# through means of my own preparation, column names are saved in a txt file
with open('columns.txt','r') as cols:
    cols = cols.read().rstrip().split("\n")
    
# if necessary check cols with a print
# print(cols)

# read in train and test
train = pd.read_csv('census/census-income.data', na_values=['?',' ',' ?',' ?',' ? ',''], names=cols)
test = pd.read_csv('census/census-income.test', na_values=['?',' ',' ?',' ?',' ? ',''], names=cols)

# train = pd.concat([train,test], ignore_index=True)

# check shapes
for ds in [train, test]:
    print("Rows: {} Columns: {}".format(*ds.shape))
```

```python
def featureEng(df):
    
    # age bins
    # buckets = np.linspace(0, 100, 15)
    buckets = np.linspace(0, 100, 15)
    labels = np.arange(14).tolist()
    df['ageBinned'] = pd.cut(df.age, bins=buckets, labels=labels)

    # tax filer
    df['is_taxfiler'] = df['tax filer stat'].apply(lambda x: 'Yes' if x != ' Nonfiler' else 'No')

    # dividends from stocks
    # if the individual has dividends totally more than 50k
    df['stockDividends50k'] = df['dividends from stocks'].apply(lambda x: 'Yes' if x > 50000 else 'No')
    df['stockDividendsPos'] = df['dividends from stocks'].apply(lambda x: 'Yes' if x > 0 else 'No')

    # stock bins // wait on it
    # train.loc['']

    # first generation US (neither parents born in US)
    firstgen = df.loc[(df['country of birth father'] != ' United-States') &
              (df['country of birth mother'] != ' United-States') &
                (df['country of birth self'] == ' United-States')].index

    df['1stGenUS'] = 'No'
    df.loc[firstgen,'1stGenUS'] = 'Yes'

    # second generation US (at least 1 parent born in US)
    secondgen = df.loc[(df['country of birth father'] == ' United-States') |
              (df['country of birth mother'] == ' United-States')].index

    df['2ndGenUS'] = 'No'
    df.loc[secondgen,'2ndGenUS'] = 'Yes'

    # gains over losses
    df['gainsOvLosses'] = df.apply(lambda row: 'No' if row['capital gains'] <= row['capital losses'] else 'Yes', axis=1)

    # fi
    df['fi_married|sex|race'] = df['marital stat'] + '|' + df['race'] + '|' + df['sex']

    # agebin * weeks worked 
#     df['fi_age*weeksworked'] = (df['ageBinned'] + 1) * df['weeks worked in year']
    
    return df

def drop_NIUs(df):
    drops = []
    for col in df.columns:
        try:
            if np.sum(df[col] == ' Not in universe') > 0:
                drops.append(col)
        except:
            pass
    return df.drop(drops, axis=1)

def simpleCleaner(df):
    # remove minors age < 16
    df = df.loc[df['age'] >= 16,:]
    
    # drop high ratios of 'Not in universe'
    df = drop_NIUs(df)
        
    # clean up the last variables with NAs
    for col in ['country of birth self','country of birth mother','country of birth father']:
        df[col] = df[col].fillna('Unknown')
        
    # target
    targ = df['target'].apply(lambda x: int(x.replace(".","").strip("+").replace(" ","")))
    df['target'] = targ.map({50000:1,-50000:0})
    
    # engineer new features
    df = featureEng(df)
    
    # final features selected
    final_features = ['fi_married|sex|race','fi_age*weeksworked','ageBinned','2ndGenUS','1stGenUS',
                      'is_taxfiler','stockDividends50k','stockDividendsPos','sex','marital stat',
                      'education','full or part time employment stat','target']
    df = df.loc[:,final_features]
    
    return df
```

```python
def combineAndPrep(train, test):
    # combine
    train_i = train.index.tolist()
    test.set_index(np.arange(len(train),len(test)+len(train)), inplace=True)
    test_i = test.index.tolist()
    combine = pd.concat([train, test], ignore_index=True)
    # clean
    combine = simpleCleaner(combine)
    combine = pd.get_dummies(combine)
    print()
    # split
    train_o = list(set(train_i) & set(combine.index.tolist()))
    test_o = list(set(test_i) & set(combine.index.tolist()))
    train = combine.loc[train_o,:]
    test = combine.loc[test_o,:]
    #return
    return train, test
```

```python
print("Shapes before:\nTrain: {}, {}\nTest: {}, {}".format(*train.shape, *test.shape))
train, test = combineAndPrep(train, test)
print("Shapes after:\nTrain: {}, {}\nTest: {}, {}".format(*train.shape, *test.shape))
```

```python
def underSample(df, targ):
    pos = df.loc[df[targ] == 1]
    neg = df.loc[df[targ] == 0]
    neg_sampled = df.sample(len(pos), replace=True, random_state=33)
    us = pd.concat([pos,neg_sampled], ignore_index=True)
    return us
```

```python
us_train = underSample(train, 'target')
print(us_train.shape)
```

```python
# ALGOS :)

kf = KFold(10, shuffle=True, random_state=33)

X, y = us_train.loc[:,us_train.columns != 'target'], us_train['target']

lr = LogisticRegression(random_state=33)
rf = RandomForestClassifier(random_state=33)
gb = GradientBoostingClassifier(random_state=33)
xg = xgb.XGBClassifier(random_state=33)

algos = [lr, rf, gb, xg]
feat_imps = {}

for algo in algos:
    name = str(algo.__class__).strip("'>").split('.')[-1]
    algo.fit(X,y)
    scores = cross_val_score(algo, X, y, scoring='recall', cv=kf)
    print("{}:{}\nMean:{}".format(name, scores, np.mean(scores)))
    preds = algo.predict(test.loc[:,test.columns != 'target'])
    print("On TEST: {}\n".format(recall_score(test['target'].values, preds)))
    try:
        feat_imps[name] = algo.feature_importances_
    except:
        feat_imps[name] = algo.coef_[0]
```

```python
plt.figure(figsize=(10,25));
plt.title("Feature importances");
legends=[]
for algo in feat_imps:
    if algo != 'LogisticRegression':
        plt.barh(X.columns, feat_imps[algo]);
        legends.append(algo)
plt.legend(legends);
```
